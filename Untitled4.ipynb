```python
corr_matrix = X.corr().abs()

# upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# index of feature columns with correlation greater than 0.95
highly_correlated = [column for column in upper.columns if any(upper[column] > 0.95)]

print(f"Highly correlated features (threshold > 0.95): {highly_correlated}")
```

```python
X.drop(columns=highly_correlated, inplace=True)
```

```python
# Fit and transforming the data using the original DataFrame
threshold = 0.01  
sel = VarianceThreshold(threshold=threshold)
sel.fit(X)

# indices of the retained features
retained_features_indices = sel.get_support()

# Index the original DataFrame columns
retained_features = X.columns[retained_features_indices]

print(f"Features retained after variance thresholding: {list(retained_features)}")
```

```python
X_scaled.shape
```

```python
#  manually reintroducing certain features
manually_selected_features = ['Fwd Packet Length Max', 'Bwd Packet Length Max']

# Checking if manually selected features are in the original DataFrame
for feature in manually_selected_features:
    if feature not in X.columns:
        print(f"Warning: Feature '{feature}' not found in the DataFrame. It won't be added.")

# Combining retained features with manually selected features that exist in the DataFrame
final_features = list(retained_features)
for feature in manually_selected_features:
    if feature in X.columns:
        final_features.append(feature)

#  X_final using the selected features x is original dataframe
X_final = X[final_features]

#  PCA
# Let's reduce to 2 components for visualization purposes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_final)

# Plot the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis') # Assuming 'y' is your target variable
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Dataset')
plt.show()
```

```python
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by each component: {explained_variance}")
```

Model Building

```python
from sklearn.model_selection import train_test_split

y_final = data['Label']  # Assuming 'Label' is the target variable

# Spliting the data into training and testing sets (e.g., 70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.3, random_state=42)

print(f"Training feature set shape: {X_train.shape}")
print(f"Test feature set shape: {X_test.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test labels shape: {y_test.shape}")
```

```python
from sklearn.svm import SVC

# Initialize and train the model
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Make predictions
y_pred_svm = svm_model.predict(X_test)

# Evaluate the model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm}")
print("Classification Report for SVM:")
print(classification_report(y_test, y_pred_svm))

# Confusion Matrix
ConfusionMatrixDisplay.from_estimator(svm_model, X_test, y_test)
plt.title("SVM Confusion Matrix")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

# Make predictions
y_pred_lr = lr_model.predict(X_test)

# Evaluate the model
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Logistic Regression Accuracy: {accuracy_lr}")
print("Classification Report for Logistic Regression:")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
ConfusionMatrixDisplay.from_estimator(lr_model, X_test, y_test)
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and train the model
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)

# Make predictions
y_pred_gnb = gnb_model.predict(X_test)

# Evaluate the model
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)
print(f"Gaussian Naive Bayes Accuracy: {accuracy_gnb}")
print("Classification Report for Gaussian Naive Bayes:")
print(classification_report(y_test, y_pred_gnb))

# Confusion Matrix
ConfusionMatrixDisplay.from_estimator(gnb_model, X_test, y_test)
plt.title("Gaussian Naive Bayes Confusion Matrix")
plt.show()
```
